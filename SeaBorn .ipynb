```python
import seaborn as sns
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

# Distplot

```python
sns.distplot(tips[['total_bill']])
```

```python
sns.distplot(tips[['tip']], kde=False, bins=30)
```

# JointPlot()

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

# Stripplot()
stripplot provide a simple way to show the value of some quantative variable across the lavel of categorical variable. 
it drow a scatter plot when varibale in categorical 

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', jitter=False, data=tips)
```

```python
sns.stripplot(x='day', y='total_bill',data=tips, jitter=False, hue='sex')
```

# swarmplot()
swarmplot() is similler to the stripplot() And uses different approach to avoid the overlaping 

```python
sns.swarmplot(x='day',  y='total_bill', data= tips, hue='sex', )
```

```python
sns.swarmplot(x='total_bill',  y='day', data= tips, hue='sex', )
```

# Boxplot
These type of plots are used to show the distribution of categorical data. A boxplot(also know as box-and-whisker plot)
show the dustribution of quantitative data in way the facilitates compiresion between variables or across levels of categorical variable   

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex')
```

# Violinplot
The  Violinplot() Draw a combinition boxplot and karnel density estimate 
A violin plot plays a similer role 

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```
